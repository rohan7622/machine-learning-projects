# K-Nearest Neighbours Classification

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,:-1]
y=df["Purchased"]
```

```python
sns.scatterplot(x="Age",y="EstimatedSalary",data=df,hue="Purchased")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
sc.fit(x)
x=pd.DataFrame(sc.transform(x),columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)*100
```

```python
knn.score(x_train,y_train)*100
```

```python
knn.predict([[-1.113206,-0.785290]])
```

```python
for i in range(1,30):
    knn1=KNeighborsClassifier(n_neighbors=i)
    knn1.fit(x_train,y_train)
    print(i,knn1.score(x_train,y_train)*100,knn1.score(x_test,y_test)*100)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=knn)
plt.show()
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=knn1)
plt.show()
```
