# standerdscaler--mean and standerd devition


```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# example data
data = {'age':[20,30,40,50,60],
        'income':[20000,30000,40000,50000,60000]}
```

```python
df = pd.DataFrame(data)
df
```

```python
# creat the standerdscaler object
scaler = StandardScaler()
```

```python
# fit and transform the data
scaled_data= scaler.fit_transform(df)
```

```python
# creat data frame scale data
scaled_df = pd.DataFrame(scaled_data,columns=df.columns)

print("Standerd Scaled Data:")

print(scaled_df)
```

# 2.minmaxscaler-range,max-min

```python
from sklearn.preprocessing import MinMaxScaler

df=pd.DataFrame(data)
df
```

```python
# create min max scaler object

minmax_scaler =MinMaxScaler()
```

```python
minmax_scaler
```

```python
# fit and transform the data
minmax_scaled_data = minmax_scaler.fit_transform(df)
```

```python
minmax_scaled_data
```

```python
# creat dataframe from scaled data
minmax_scaled_df = pd.DataFrame(minmax_scaled_data, columns=df.columns)
print("Min-Max Scaled Data:")
print(minmax_scaled_df)
```

# 3 robust scaler - --outlier-- median interquartile

```python
from sklearn.preprocessing import RobustScaler

# example data
df=pd.DataFrame(data)
df
```

```python
# creat robustscaler object
robust_scaler = RobustScaler()
```

```python
# fit and transform the data
robust_scaled_data = robust_scaler.fit_transform(df)
```

```python
# creat dataframe from scaled data
robust_scaled_df = pd.DataFrame(robust_scaled_data,columns=df.columns)


print("Robust Scaled Data:")
print(robust_scaled_df)
```
