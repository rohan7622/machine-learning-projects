### Order Encoding Using two techique
* 1. SKlearn and
* 2. Map Funtion

```python
# 1 sklearn
import pandas as pd
```

```python
df = pd.DataFrame({"size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
df.head(3)
```

```python
# Two dimension data set so we use double Brakcet
ord_data=[["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe= OrdinalEncoder(categories = ord_data)
oe.fit(df[["size"]])
```

```python
df["size_en"] = oe.transform(df[["size"]])
df
```

```python
# 2. Map Funtion
ord_data1 = {"s":5,"m":6,"l":7,"xl":8}
```

```python
df["size"].map(ord_data1)
```

```python
df["size_en-map"] = df["size"].map(ord_data1)
df
```
