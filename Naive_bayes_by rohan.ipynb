```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv("placement-dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="cgpa",y="iq",data=df,hue="placement")
plt.show()
```

```python
x=df.iloc[:,:-1]
y=df["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
sns.kdeplot(df["cgpa"])
plt.show()
```

```python
sns.kdeplot(df["iq"])
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)*100,model.score(x_train,y_train)*100
```

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100,mnb.score(x_train,y_train)*100
```

```python
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100,bnb.score(x_train,y_train)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=model)
plt.show()
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```
