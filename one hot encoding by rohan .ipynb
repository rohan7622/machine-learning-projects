## One Hot Encoding  and Dummy Variables (is performed on Cartegorical data )
#### Encoding in machine learning is the process of converting categorical data into numerical values so that machine learning algorithms can process it. 

#### Married = Yes /NO -0 /1
#### Gender  = Male /Female - 0 /1
##### When limited amount of data is present in the data set then we use one hot encodin

```python
import pandas as pd
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head()
```

```python
#filling the missing values
dataset.isnull().sum()
```

```python
# gender = 13
# married = 3
```

```python
# new
for i in dataset.select_dtypes(include='object').columns:
    dataset [i] = dataset[i].fillna(dataset[i].mode()[0])
```

```python
dataset.isnull().sum()
```

```python
# fill the data in married column
dataset["Married"] = dataset["Married"].fillna(dataset["Married"].mode()[0])
```

```python
dataset.isnull().sum()
```

```python
# One hot encoding we have two option pandas - get.dummies(), sckitlearn _ Onehotencoder
```

```python
# first we seprate the data gender and Married
en_data =dataset[["Gender","Married"]]
en_data
```

```python
pd.get_dummies(en_data)
```

```python
pd.get_dummies(en_data).info()
```

```python
# we get the data in boolen data type 
# so we have to convert boolen datatype to numerical data type
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
ohe.fit_transform(en_data)
```

```python
# sparse matrix means data is fill with 0 and 1
```

```python
# to covert the data into array format
ohe = OneHotEncoder()
ohe.fit_transform(en_data).toarray()
```

```python
# to covert the data into dataframe
ar= ohe.fit_transform(en_data).toarray()
pd.DataFrame(ar,columns=["Gender_Female","Gender_Male","Married_No","Married_Yes" ])
```

```python
# Now in encoding we have given only two columns but after encoding there are Four coloumns Gender_ female, male, married no, yes
# Our data is now extend so we make it small use (drop ="First")
```

```python
ohe = OneHotEncoder(drop="first")
ar =ohe.fit_transform(en_data).toarray()
ar
```

```python
pd.DataFrame(ar,columns=["Gender_Male","Married_Yes" ])
```
