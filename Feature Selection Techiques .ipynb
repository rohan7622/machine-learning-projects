```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
dataset = pd.read_csv("diabetes.csv")
```

```python
dataset.head(3)
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Outcome"]
```

```python
x.shape
```

```python
#iloc[:, :-1] selects all rows (:) and all columns except the last one (:-1).
# This is used to assign the feature columns (input variables) to x.
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr =LogisticRegression()
```

```python
fs= SequentialFeatureSelector(lr,k_features=5,forward=True)
fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```
